```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# file_precios_leche = 'precio-promedio-del-litro-de-leche-2011-2017.utf8.csv'
# precios_df = pd.read_csv(file_precios_leche)

file_precios_leche = 'precios.csv'
precios_data = open(file_precios_leche).readlines()

fechas = []
precios = []

for line in precios_data:
    line = line.strip()
    ano,mes,precio = line.split(',')
    fechas.append(ano + '_' + mes)
    precios.append(float(precio)/1000.0)
    
fechas,precios
```

```python
plt.rcParams['figure.figsize'] = [40,20]

plt.style.use('grayscale') # dark_background
plt.plot(precios, color="blue",linewidth=15)
#plt.xticks(rotation=90)
#plt.xticks(range(75), fechas)

#ax = plt.gca()
#ax.set_facecolor('chartreuse')

ppo_ano = [0,12,24,36,48,60,72]

for ano in ppo_ano:
    plt.annotate(fechas[ano][0:4] + ": " + str(precios[ano]), (ano,precios[ano]), textcoords="offset points", xytext=(0,10), ha='center')

#plt.savefig('precio_leche.png')
```

```python
natalidad_data = open('natalidad.csv').readlines()

fechas_natalidad = []
natalidad = []

for line in natalidad_data:
    line = line.strip()
    ano,tasa,fertilidad = line.split(',')
    fechas_natalidad.append(ano)
    natalidad.append(float(tasa))
    
fechas_natalidad,natalidad
```

```python
plt.rcParams['figure.figsize'] = [40,20]

plt.style.use('grayscale')
plt.gca().yaxis.set_label_position("right")
plt.plot(fechas_natalidad,natalidad)
# plt.savefig('natalidad.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()
t = np.arange(17, 18.0, 0.01)
s1 = np.exp(t)
ax1.plot(t, s1, 'b-')
ax1.set_xlabel('time (s)')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('exp', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
s2 = np.arange(17, 18.0, 0.01)
ax2.plot(t, s2, 'r.')
ax2.set_ylabel('sin', color='r')
ax2.tick_params('y', colors='r')

fig.tight_layout()
```
